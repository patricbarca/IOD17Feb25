````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''

    N = len(X)
    print(N)
    sumx = np.sum(X)
    sumy = np.sum(Y)
    sumxy = np.sum(np.array(X)*np.array(Y))
    sumxexp2 = np.sum(np.array(X)**2)

    #print(sumx)
    #print(sumy)
    #print(sumxy)
    
    #print(sumxexp2)

    m = (N*sumxy-sumx*sumy)/(N*sumxexp2-(sumx)**2)
    #print(m)
    b = (sumy-m*sumx)/N
    #print(b)
    return m, b

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X,Y)
yfit = [b+m*xi for xi in X]
plt.plot(X,yfit)

plt.show()
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
# ANSWER


stats.probplot(x, dist="norm",plot=ax1)
ax1.set_title("Q-Q Plot of Original Data")

x_transformed, lambda_val = stats.boxcox(x)

ax2 = fig.add_subplot(212)
stats.probplot(x_transformed, dist="norm", plot=ax2)
ax2.set_title(f"Q-Q Plot After Box-Cox Transformation (={lambda_val:.2f})")

plt.tight_layout()
plt.show()
````

# 13. [BONUS] Calculate Mean Confidence Interval

Check the following methods:

Expected output:

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m = np.mean(a)  # Compute mean
  se = scipy.stats.sem(a)  # Compute standard error of the mean (SEM)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
