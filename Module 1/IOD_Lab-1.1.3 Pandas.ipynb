# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
# A:
n, bin, patches = plt.hist(df['age'],bins=40,edgecolor='black')
plt.xlabel('Age')
# plt.xticks(range(min(df['age']),max(df['age'])+1))
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
````

#### 4.B BoxPlot of `age`

````python
# A:
plt.boxplot(df['age'], vert=True)
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.show()
````

````python
plt.boxplot(df['age'], vert=False)
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
# A:
plt.scatter(df['age'],df['length_of_service'],color='purple')
plt.xlabel('Age')
plt.ylabel('Length of service')
plt.title('Age vs Length of service')
plt.show()
````
